```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
suv_car_df = pd.read_csv('suv_data.csv')
```

```python
# Display the first few rows of the dataframe
suv_car_df.head()
```

```python
# Display the shape of the dataframe
suv_car_df.shape
```

```python
# Display the 'Purchased' column
suv_car_df['Purchased']
```

# Extracting independent and dependent variable

```python
# Extract independent and dependent variables
X = suv_car_df.iloc[:,[2,3]]
y = suv_car_df.iloc[:,4]
```

```python
# Display the shape of the independent variable X
X.shape
```

```python
# Display the shape of the dependent variable y
y.shape
```

# Train Test Split

```python
# Import train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
# Display the shape of the training set for X
X_train.shape
```

```python
# Display the shape of the testing set for X
X_test.shape
```

# Scalling

```python
# Scale the independent variables
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
#
```

```python
# Display the scaled testing set for X
X_test
```

# Train Model

```python
# Train the Logistic Regression model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# Make predictions on the training and testing sets
train_y_pred = model.predict(X_train)
test_y_pred = model.predict(X_test)
```

# Accuracy Score of model


```python
# Calculate and print accuracy scores
from sklearn.metrics import accuracy_score
print('Train set accuracy is :', accuracy_score(y_train,train_y_pred))
print('Test set accuracy is :', accuracy_score(y_test,test_y_pred))
```

# Confusion Metrics

```python
# Calculate the confusion matrix
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test,test_y_pred)
```

```python
# Display the confusion matrix
cf
```

# Let's Plot the confusion matrix

```python
# Plot the confusion matrix
import seaborn as sns
sns.heatmap(cf,annot=True)
```
