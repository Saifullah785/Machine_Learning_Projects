```python
# importing necessary libraries

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset using pandas 

credit_card_data = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
credit_card_data.head()
```

```python
credit_card_data.sample()
```

```python
# dataset information

credit_card_data.info()
```

```python
# checking the number of missing values in each column

credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction & froudulent transaction

credit_card_data['Class'].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data['Class']==1]
```

```python
fraud['Class']
```

```python
# statistical measures of the data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions

credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample (n=492)
```

```python
new_df = pd.concat([legit_sample,fraud],axis= 0)
```

```python
new_df
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X = new_df.drop(columns= 'Class', axis =1)
Y= new_df['Class']
```

```python
# spliting the dataset in testing and training sets

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

# **Model Training**

**Logistic Regression**

```python
# creating a object Logistic regression class

model = LogisticRegression()
```

```python
# training the logistic Regression model with Training Data

model.fit(X_train, Y_train)

# accuracy on training data

X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training dat :',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```
