**Define a function that tokenizes and stems the text of the product tiles and description using the NLTK library**

```python
# Import the SnowballStemmer from NLTK for stemming words
from nltk.stem.snowball import SnowballStemmer
# Initialize the SnowballStemmer for English language
stemmer = SnowballStemmer("english")

# Define a function to tokenize and stem the input text
def tokenize_stem(text):
    # Tokenize the text into words and convert to lowercase
    tokens = nltk.word_tokenize(text.lower(), language='english')
    # Stem each token using the SnowballStemmer
    stem = [stemmer.stem(w) for w in tokens]
    # Join the stemmed tokens back into a string
    return " ".join(stem)
```

**Create a new column in the DataFrame that contains the stemmed tokens of the product titles and descriptions.**

```python
# Apply the tokenize_stem function to the combined 'Title' and 'Description' columns
# and create a new column 'stemmed_tokens' with the results
amzon_df['stemmed_tokens'] = amzon_df.apply(lambda row: tokenize_stem(row['Title'] + ' ' + row['Description']), axis=1)
# Display the first few rows of the DataFrame with the new 'stemmed_tokens' column
amzon_df.head()
```

# **what is cosine similarity**

Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them.

In the context of natural language processing, it is commonly used to measure the similarity between two text documents.

The cosine similarity score ranges from 0 to 1, where 0 means no similarity and 1 means identical.

The score is calculated based on the frequency of common words or terms in the two documents.

The higher the cosine similarity score, the more similar the documents are.

# **Define a function that takes a query string and returns a DataFrame with the top 10 most relevent products based on the cosine similarity between the query string and the product titles and description.**

```python
# Define a function to search for products based on a query string
def search_product(query):

    # Tokenize and stem the query string
    stemmed_query = tokenize_stem(query)

    # Transform the stemmed query into a TF-IDF vector using the fitted vectorizer
    query_vector = vectorizer.transform([stemmed_query])

    # Calculate the cosine similarity between the query vector and the TF-IDF matrix
    # This gives a similarity score for each product in the DataFrame
    amzon_df['similarity'] = cosine_similarity(query_vector, tfidf_matrix).flatten()

    # Sort the DataFrame by similarity in descending order and select the top 10 products
    # Then, select only the 'Title', 'Description', and 'Category' columns
    res = amzon_df.sort_values(by=['similarity'],ascending=False).head(10)[['Title','Description','Category']]

    # Return the DataFrame containing the top 10 most relevant products
    return res
```
