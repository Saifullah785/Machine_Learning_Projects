# **Grouping by 'citizenship' to find out how many incidents involve each nationality**

```python
# Group by 'citizenship', count the occurrences, and reset the index to create a DataFrame

df.groupby('citizenship').size().reset_index(name='incident_count')
```

```python
# Define a function to group by 'citizenship' and count incidents

def nationality_incident(df):
  nationality_incident = df.groupby('citizenship').size().reset_index(name='incident_count')
  return nationality_incident

nationality_incident(df)
```
